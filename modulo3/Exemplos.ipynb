## Matplotlib

```python
import random

import matplotlib.pyplot as plt
import numpy as np

random.seed(42)
```

```python
# 1) Gráfico de barras
categorias = ['A', 'B', 'C', 'D']
valores = [23, 17, 35, 29]

plt.figure(figsize=(8, 4))
plt.bar(categorias, valores, color=['#4C78A8', '#F58518', '#54A24B', '#E45756'])
plt.title('Exemplo: Gráfico de Barras')
plt.xlabel('Categoria')
plt.ylabel('Valor')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# 2) Gráfico de linhas
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun']
vendas = [120, 135, 150, 145, 160, 175]

plt.figure(figsize=(8, 4))
plt.plot(meses, vendas, marker='o', linewidth=2, color='#4C78A8')
plt.title('Exemplo: Gráfico de Linhas')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.grid(alpha=0.3)
plt.show()
```

```python
# 3) Boxplot
notas_turma_a = np.random.normal(loc=7.0, scale=1.0, size=30)
notas_turma_b = np.random.normal(loc=6.5, scale=1.2, size=30)
notas_turma_c = np.random.normal(loc=8.0, scale=0.8, size=30)

plt.figure(figsize=(8, 4))
plt.boxplot([notas_turma_a, notas_turma_b, notas_turma_c], labels=['Turma A', 'Turma B', 'Turma C'])
plt.title('Exemplo: Boxplot de Notas por Turma')
plt.ylabel('Nota')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# 4) Histograma
idades = np.random.normal(loc=30, scale=8, size=200)

plt.figure(figsize=(8, 4))
plt.hist(idades, bins=15, color='#54A24B', edgecolor='black', alpha=0.8)
plt.title('Exemplo: Histograma de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# 5) Gráfico de setores (pizza)
segmentos = ['Produto A', 'Produto B', 'Produto C', 'Produto D']
participacao = [40, 25, 20, 15]

plt.figure(figsize=(6, 6))
plt.pie(participacao, labels=segmentos, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'white'})
plt.title('Exemplo: Gráfico de Setores')
plt.axis('equal')
plt.show()
```

## Exemplos dos mesmos gráficos com Seaborn

A seguir, os mesmos tipos de gráfico usando `seaborn`: **barras**, **linhas**, **boxplot**, **histograma** e **setores (pizza)**.

```python
import seaborn as sns
import pandas as pd

sns.set_theme(style='whitegrid')
```

```python
# 1) Barras com Seaborn
df_barras = pd.DataFrame({
    'Categoria': ['A', 'B', 'C', 'D'],
    'Valor': [23, 17, 35, 29]
})

plt.figure(figsize=(8, 4))
sns.barplot(data=df_barras, x='Categoria', y='Valor', palette='Set2')
plt.title('Seaborn: Gráfico de Barras')
plt.xlabel('Categoria')
plt.ylabel('Valor')
plt.show()
```

```python
# 2) Linhas com Seaborn
df_linhas = pd.DataFrame({
    'Mês': ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun'],
    'Vendas': [120, 135, 150, 145, 160, 175]
})

plt.figure(figsize=(8, 4))
sns.lineplot(data=df_linhas, x='Mês', y='Vendas', marker='o', linewidth=2, color='#4C78A8')
plt.title('Seaborn: Gráfico de Linhas')
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.show()
```

```python
# 3) Boxplot com Seaborn
df_box = pd.DataFrame({
    'Turma': ['Turma A'] * 30 + ['Turma B'] * 30 + ['Turma C'] * 30,
    'Nota': np.concatenate([
        np.random.normal(loc=7.0, scale=1.0, size=30),
        np.random.normal(loc=6.5, scale=1.2, size=30),
        np.random.normal(loc=8.0, scale=0.8, size=30)
    ])
})

plt.figure(figsize=(8, 4))
sns.boxplot(data=df_box, x='Turma', y='Nota', palette='pastel')
plt.title('Seaborn: Boxplot de Notas por Turma')
plt.xlabel('Turma')
plt.ylabel('Nota')
plt.show()
```

```python
# 4) Histograma com Seaborn
idades_sns = np.random.normal(loc=30, scale=8, size=200)

plt.figure(figsize=(8, 4))
sns.histplot(idades_sns, bins=15, kde=True, color='#54A24B')
plt.title('Seaborn: Histograma de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# 5) Setores (pizza) com paleta Seaborn
segmentos_sns = ['Produto A', 'Produto B', 'Produto C', 'Produto D']
participacao_sns = [40, 25, 20, 15]
cores = sns.color_palette('Set2', n_colors=len(segmentos_sns))

plt.figure(figsize=(6, 6))
plt.pie(
    participacao_sns,
    labels=segmentos_sns,
    autopct='%1.1f%%',
    startangle=90,
    colors=cores,
    wedgeprops={'edgecolor': 'white'}
)
plt.title('Setores (pizza) com paleta Seaborn')
plt.axis('equal')
plt.show()
```

## Exemplos Seaborn com conjuntos de dados reais

Nesta seção, usamos datasets reais disponibilizados pelo próprio Seaborn (`tips`, `penguins`, `flights` e `iris`).

```python
# Carregando datasets reais do Seaborn
tips = sns.load_dataset('tips')
penguins = sns.load_dataset('penguins').dropna()
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')

tips.head()
```

```python
# 1) Barras: conta média por dia (tips)
plt.figure(figsize=(8, 4))
sns.barplot(data=tips, x='day', y='total_bill', estimator='mean', errorbar=None, palette='Set2')
plt.title('Tips: Média da conta por dia')
plt.xlabel('Dia da semana')
plt.ylabel('Conta média (USD)')
plt.show()
```

```python
# 2) Linhas: passageiros por mês e ano (flights)
plt.figure(figsize=(10, 5))
sns.lineplot(data=flights, x='month', y='passengers', hue='year', legend=False, palette='viridis')
plt.title('Flights: Passageiros por mês (cada linha = um ano)')
plt.xlabel('Mês')
plt.ylabel('Número de passageiros')
plt.show()
```

```python
# 3) Boxplot: massa corporal por espécie (penguins)
plt.figure(figsize=(8, 4))
sns.boxplot(data=penguins, x='species', y='body_mass_g', palette='pastel')
plt.title('Penguins: Distribuição de massa corporal por espécie')
plt.xlabel('Espécie')
plt.ylabel('Massa corporal (g)')
plt.show()
```

```python
# 4) Histograma: distribuição do valor da conta (tips)
plt.figure(figsize=(8, 4))
sns.histplot(data=tips, x='total_bill', bins=20, kde=True, color='#4C78A8')
plt.title('Tips: Distribuição do valor da conta')
plt.xlabel('Valor da conta (USD)')
plt.ylabel('Frequência')
plt.show()
```

```python
# 5) Setores (pizza): proporção de clientes por horário (tips)
contagem_turno = tips['time'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(
    contagem_turno.values,
    labels=contagem_turno.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('Set2', n_colors=len(contagem_turno)),
    wedgeprops={'edgecolor': 'white'}
)
plt.title('Tips: Proporção de clientes por horário')
plt.axis('equal')
plt.show()
```
