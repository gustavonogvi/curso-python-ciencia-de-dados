```python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela"],
    "idade": [23, 35, 29, 41],
    "nota": [8.5, 6.0, 7.5, 9.0]
}

df = pd.DataFrame(dados)
df
```

---
### Exercício 1 — Seleção de Dados

```python
print(df["nome"])
print()
print(df[["nome", "nota"]])
```

---
### Exercício 2 — Filtro

```python
print(df[df["nota"] > 7])
print()
print(df[df["idade"] > 30])
```

---
### Exercício 3 — Nova Coluna

```python
df["status"] = df["nota"].apply(lambda n: "Aprovado" if n >= 7 else "Reprovado")
print(df)
```

---
### Exercício 4 — Ordenação

```python
print(df.sort_values(by="idade"))
print()
print(df.sort_values(by="nota", ascending=False))
```

---
### Exercício 5 — Estatísticas

```python
print(f"Média das idades: {df['idade'].mean():.1f}")
print(f"Maior nota:       {df['nota'].max()}")
print(f"Menor nota:       {df['nota'].min()}")
```

---
### Exercício 6 — Limpeza de Dados

```python
import pandas as pd

dados_sujos = {
    "nome": ["Ana", "Bruno", "Carlos", "Ana"],
    "idade": [23, None, 29, 23],
    "nota": [8.5, 7.0, None, 8.5]
}

df_sujo = pd.DataFrame(dados_sujos)

print(df_sujo.isnull().sum())
print(f"Duplicatas: {df_sujo.duplicated().sum()}")

df_sujo["nota"] = df_sujo["nota"].fillna(df_sujo["nota"].mean())
df_sujo = df_sujo.drop_duplicates()

print()
print(df_sujo)
```
