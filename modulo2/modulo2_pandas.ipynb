## 1. Introdução ao Pandas

Pandas é a principal biblioteca para análise de dados em Python. Ela oferece duas estruturas fundamentais:

| Estrutura | Descrição | Quando usar |
|-----------|-----------|-------------|
| `Series` | Coluna única com índice | Uma variável |
| `DataFrame` | Tabela com linhas e colunas | Dados completos |

````python
# pd e np são as convenções universais — todo mundo usa assim
import pandas as pd
import numpy as np
````

````python
# Series é basicamente uma lista com índice
# raramente criamos direto assim, mas é bom entender a estrutura
notas = pd.Series([8.5, 7.0, 9.2, 6.5])
print(notas)
print("Tipo:", type(notas))
````

````python
# na prática, o DataFrame vem de um CSV ou banco de dados
# aqui criamos manualmente só pra ver como funciona
dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "idade": [23, 35, 29, 41, 27],
    "nota": [8.5, 6.0, 7.5, 9.0, 5.5]
}

df = pd.DataFrame(dados)
df
````

````python
# shape e dtypes são os primeiros que você checa quando recebe um dataset novo
print("Shape (linhas, colunas):", df.shape)
print()
print("Colunas:", df.columns.tolist())
print()
print("Tipos de dados:")
print(df.dtypes)
````

## 2. Leitura e Escrita de Dados

Pandas suporta os principais formatos de arquivo usados em dados.

| Formato | Leitura | Escrita |
|---------|---------|--------|
| CSV | `read_csv()` | `to_csv()` |
| Excel | `read_excel()` | `to_excel()` |
| JSON | `read_json()` | `to_json()` |

````python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "idade": [23, 35, 29, 41, 27],
    "nota": [8.5, 6.0, 7.5, 9.0, 5.5]
}

df = pd.DataFrame(dados)

# index=False evita criar uma coluna extra com 0,1,2,3... no arquivo
df.to_csv("turma.csv", index=False)
print("CSV salvo.")

df_csv = pd.read_csv("turma.csv")
print(df_csv)
````

````python
# orient='records' salva cada linha como um objeto — é o formato mais comum de JSON
# force_ascii=False garante que acentos não virem caracteres estranhos
df.to_json("turma.json", orient="records", force_ascii=False)
print("JSON salvo.")

df_json = pd.read_json("turma.json")
print(df_json)
````

````python
# esses quatro comandos são o primeiro que você roda ao receber qualquer dataset
print("--- head() ---")
print(df.head())

print("\n--- tail() ---")
print(df.tail())

print("\n--- info() ---")
df.info()

print("\n--- describe() ---")
print(df.describe())
````

## 3. Limpeza de Dados

Dados reais raramente chegam prontos. Precisamos identificar e corrigir problemas antes de analisar.

| Problema | Detecção | Solução |
|----------|----------|---------|
| Valores nulos | `isnull().sum()` | `dropna()` ou `fillna()` |
| Duplicatas | `duplicated().sum()` | `drop_duplicates()` |
| Tipo errado | `dtypes` | `astype()` |

````python
import pandas as pd

# esse tipo de dado sujo é muito comum quando o CSV vem de planilhas
# repare que idade está como string — isso acontece quando a coluna tem célula vazia
dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Ana", "Eva"],
    "idade": ["23", None, "29", "23", "31"],
    "nota": [8.5, 7.0, None, 8.5, 6.0]
}

df = pd.DataFrame(dados)
print(df)
print()
df.info()
````

````python
# sempre cheque nulos e duplicatas antes de qualquer análise
print("Nulos por coluna:")
print(df.isnull().sum())

print(f"\nLinhas duplicadas: {df.duplicated().sum()}")
````

````python
# substituímos nota nula pela média — melhor do que perder a linha inteira
df["nota"] = df["nota"].fillna(df["nota"].mean())

# idade não tem como adivinhar, então removemos a linha
df = df.dropna()

df = df.drop_duplicates()

# agora que os nulos sumiram, podemos converter o tipo com segurança
df["idade"] = df["idade"].astype(int)

print("Depois da limpeza:")
print(df)
print()
df.info()
````

## 4. Seleção e Manipulação de Dados

### Selecionando colunas

```python
df["nome"]            # uma coluna → Series
df[["nome", "nota"]]  # múltiplas colunas → DataFrame
```

### Selecionando linhas com `loc` e `iloc`

| Método | Acessa por | Exemplo |
|--------|------------|---------|
| `loc` | rótulo/condição | `df.loc[0]`, `df.loc[df["nota"] > 7]` |
| `iloc` | posição numérica | `df.iloc[0]`, `df.iloc[1:3]` |

````python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "idade": [23, 35, 29, 41, 27],
    "nota": [8.5, 6.0, 7.5, 9.0, 5.5]
}

df = pd.DataFrame(dados)

# colchete simples → Series; colchete duplo → DataFrame
print(df["nome"])
print()
print(df[["nome", "nota"]])
````

````python
# loc usa o rótulo do índice — no nosso caso é numérico, mas poderia ser texto
# atenção: loc é inclusivo nos dois lados (1:3 pega 1, 2 e 3)
print(df.loc[0])       # linha de índice 0
print()
print(df.loc[1:3])     # linhas 1, 2 e 3
print()
print(df.loc[df["nota"] >= 7])  # filtro por condição
````

````python
# iloc usa posição — funciona igual ao fatiamento de listas
# iloc[1:3] pega posições 1 e 2 (3 fica de fora)
print(df.iloc[0])      # primeira linha
print()
print(df.iloc[1:3])    # posições 1 e 2
print()
print(df.iloc[-2:])    # últimas 2 linhas
````

````python
# apply com lambda é a forma mais direta de criar coluna condicional
df["status"] = df["nota"].apply(lambda n: "Aprovado" if n >= 7 else "Reprovado")

df_ord = df.sort_values(by="nota", ascending=False)
print(df_ord)

# groupby agrega os dados por categoria — muito útil para comparar grupos
print()
print(df.groupby("status")["nota"].mean())
````

## 5. Estatística Descritiva com Pandas e NumPy

Antes de qualquer análise, é essencial entender a distribuição dos dados.

| Medida | Pandas | NumPy |
|--------|--------|-------|
| Média | `mean()` | `np.mean()` |
| Mediana | `median()` | `np.median()` |
| Moda | `mode()` | — |
| Desvio padrão | `std()` | `np.std()` |
| Variância | `var()` | `np.var()` |

````python
import pandas as pd
import numpy as np

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "idade": [23, 35, 29, 41, 27],
    "nota": [8.5, 6.0, 7.5, 9.0, 5.5]
}

df = pd.DataFrame(dados)

# média é influenciada por outliers — compare sempre com a mediana
print(f"Média:   {df['nota'].mean():.2f}")
print(f"Mediana: {df['nota'].median():.2f}")
print(f"Moda:    {df['nota'].mode().values}")
````

````python
# desvio padrão alto = notas bem espalhadas; baixo = turma homogênea
print(f"Desvio padrão: {df['nota'].std():.2f}")
print(f"Variância:     {df['nota'].var():.2f}")
print(f"Mínimo:        {df['nota'].min()}")
print(f"Máximo:        {df['nota'].max()}")
print(f"Amplitude:     {df['nota'].max() - df['nota'].min():.2f}")
````

````python
# NumPy trabalha com arrays — mais rápido que Pandas em operações matemáticas puras
notas = np.array(df["nota"])

print(f"Média:         {np.mean(notas):.2f}")
print(f"Mediana:       {np.median(notas):.2f}")
print(f"Desvio padrão: {np.std(notas):.2f}")
print(f"Variância:     {np.var(notas):.2f}")

# normalização coloca os valores em escala 0–1 sem perder a proporção entre eles
notas_norm = (notas - notas.min()) / (notas.max() - notas.min())
print("\nNotas normalizadas:", notas_norm.round(2))
````

````python
# describe() entrega tudo de uma vez — é sempre o primeiro passo numa análise
print(df.describe())
````

## 6. Exemplo Prático — Fluxo Completo

Vamos simular o ciclo completo: carregar, limpar, manipular e analisar.

````python
import pandas as pd
import numpy as np

# dados com os problemas típicos de um CSV real
dados = {
    "nome":   ["Ana", "Bruno", "Carlos", "Ana",  "Eva",   "Felipe"],
    "idade":  ["23",  "35",    "29",     "23",   None,    "31"],
    "nota":   [8.5,   7.0,    None,      8.5,    6.0,     9.5]
}

df = pd.DataFrame(dados)
print(df)
````

````python
df["nota"] = df["nota"].fillna(df["nota"].mean())
df = df.dropna()           # Eva não tem idade — sem como recuperar, remove
df = df.drop_duplicates()  # Ana estava duplicada
df["idade"] = df["idade"].astype(int)

print(df)
````

````python
df["status"] = df["nota"].apply(lambda n: "Aprovado" if n >= 7 else "Reprovado")

# reset_index(drop=True) renumera as linhas após remover duplicatas
df = df.sort_values(by="nota", ascending=False).reset_index(drop=True)

print(df)
````

````python
print(f"Média:         {df['nota'].mean():.2f}")
print(f"Mediana:       {df['nota'].median():.2f}")
print(f"Desvio padrão: {df['nota'].std():.2f}")
print()
print(df["status"].value_counts())
print()
print(df.groupby("status")["nota"].mean().round(2))
````
