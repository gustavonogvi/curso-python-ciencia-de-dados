## 1. Introdução ao Pandas

Pandas é a principal biblioteca para análise de dados em Python. Ela oferece duas estruturas fundamentais:

| Estrutura | Descrição | Quando usar |
|-----------|-----------|-------------|
| `Series` | Coluna única com índice | Uma variável |
| `DataFrame` | Tabela com linhas e colunas | Dados completos |

````python
import pandas as pd
import numpy as np
````

````python
# Series — uma única coluna com índice automático
notas = pd.Series([8.5, 7.0, 9.2, 6.5])
print(notas)
print("Tipo:", type(notas))
````

````python
# DataFrame — tabela criada a partir de um dicionário
dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "idade": [23, 35, 29, 41, 27],
    "nota": [8.5, 6.0, 7.5, 9.0, 5.5]
}

df = pd.DataFrame(dados)
df
````

````python
# Informações básicas do DataFrame
print("Shape (linhas, colunas):", df.shape)
print()
print("Colunas:", df.columns.tolist())
print()
print("Tipos de dados:")
print(df.dtypes)
````

## 2. Leitura e Escrita de Dados

Pandas suporta os principais formatos de arquivo usados em dados.

| Formato | Leitura | Escrita |
|---------|---------|--------|
| CSV | `read_csv()` | `to_csv()` |
| Excel | `read_excel()` | `to_excel()` |
| JSON | `read_json()` | `to_json()` |

````python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "idade": [23, 35, 29, 41, 27],
    "nota": [8.5, 6.0, 7.5, 9.0, 5.5]
}

df = pd.DataFrame(dados)

# Salvando como CSV
df.to_csv("turma.csv", index=False)  # index=False evita salvar a coluna de índice
print("CSV salvo.")

# Lendo de volta
df_csv = pd.read_csv("turma.csv")
print(df_csv)
````

````python
# Salvando como JSON e lendo de volta
df.to_json("turma.json", orient="records", force_ascii=False)
print("JSON salvo.")

df_json = pd.read_json("turma.json")
print(df_json)
````

````python
# Explorando o dataset
print("--- head() — primeiras 5 linhas ---")
print(df.head())

print("\n--- tail() — últimas 5 linhas ---")
print(df.tail())

print("\n--- info() — tipos e nulos ---")
df.info()

print("\n--- describe() — estatísticas ---")
print(df.describe())
````

## 3. Limpeza de Dados

Dados reais raramente chegam prontos. Precisamos identificar e corrigir problemas antes de analisar.

| Problema | Detecção | Solução |
|----------|----------|---------|
| Valores nulos | `isnull().sum()` | `dropna()` ou `fillna()` |
| Duplicatas | `duplicated().sum()` | `drop_duplicates()` |
| Tipo errado | `dtypes` | `astype()` |

````python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Ana", "Eva"],
    "idade": ["23", None, "29", "23", "31"],   # idade veio como string
    "nota": [8.5, 7.0, None, 8.5, 6.0]
}

df = pd.DataFrame(dados)
print("DataFrame original:")
print(df)
print()
df.info()
````

````python
# 1. Verificando nulos
print("Valores nulos por coluna:")
print(df.isnull().sum())

# 2. Verificando duplicatas
print(f"\nLinhas duplicadas: {df.duplicated().sum()}")
````

````python
# 3. Corrigindo os problemas

# Preenche nota nula com a média
df["nota"] = df["nota"].fillna(df["nota"].mean())

# Remove linhas com valores nulos restantes (idade)
df = df.dropna()

# Remove duplicatas
df = df.drop_duplicates()

# Converte tipo da coluna idade de str para int
df["idade"] = df["idade"].astype(int)

print("DataFrame limpo:")
print(df)
print()
df.info()
````

## 4. Seleção e Manipulação de Dados

### Selecionando colunas

```python
df["nome"]            # uma coluna → Series
df[["nome", "nota"]]  # múltiplas colunas → DataFrame
```

### Selecionando linhas com `loc` e `iloc`

| Método | Acessa por | Exemplo |
|--------|------------|---------|
| `loc` | rótulo/condição | `df.loc[0]`, `df.loc[df["nota"] > 7]` |
| `iloc` | posição numérica | `df.iloc[0]`, `df.iloc[1:3]` |

````python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "idade": [23, 35, 29, 41, 27],
    "nota": [8.5, 6.0, 7.5, 9.0, 5.5]
}

df = pd.DataFrame(dados)

# Seleção de colunas
print("Coluna nome:")
print(df["nome"])

print("\nColunas nome e nota:")
print(df[["nome", "nota"]])
````

````python
# loc — acessa por rótulo ou condição
print("Linha de índice 0:")
print(df.loc[0])

print("\nLinhas 1 a 3 (loc — inclusivo):")
print(df.loc[1:3])

print("\nFiltro com loc — nota >= 7:")
print(df.loc[df["nota"] >= 7])
````

````python
# iloc — acessa por posição numérica (como listas Python)
print("Primeira linha (posição 0):")
print(df.iloc[0])

print("\nLinhas 1 a 3 (iloc — exclusivo no fim):")
print(df.iloc[1:3])

print("\nÚltimas 2 linhas:")
print(df.iloc[-2:])
````

````python
# Filtro, nova coluna, ordenação e groupby

# Filtro simples
aprovados = df[df["nota"] >= 7]
print("Aprovados:")
print(aprovados)

# Nova coluna
df["status"] = df["nota"].apply(lambda n: "Aprovado" if n >= 7 else "Reprovado")

# Ordenação
df_ord = df.sort_values(by="nota", ascending=False)
print("\nOrdenado por nota:")
print(df_ord)

# Agrupamento
print("\nMédia de nota por status:")
print(df.groupby("status")["nota"].mean())
````

## 5. Estatística Descritiva com Pandas e NumPy

Antes de qualquer análise, é essencial entender a distribuição dos dados.

| Medida | Pandas | NumPy |
|--------|--------|-------|
| Média | `mean()` | `np.mean()` |
| Mediana | `median()` | `np.median()` |
| Moda | `mode()` | — |
| Desvio padrão | `std()` | `np.std()` |
| Variância | `var()` | `np.var()` |

````python
import pandas as pd
import numpy as np

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "idade": [23, 35, 29, 41, 27],
    "nota": [8.5, 6.0, 7.5, 9.0, 5.5]
}

df = pd.DataFrame(dados)

# Medidas de tendência central
print("=== Medidas de Tendência Central ===")
print(f"Média:   {df['nota'].mean():.2f}")
print(f"Mediana: {df['nota'].median():.2f}")
print(f"Moda:    {df['nota'].mode().values}")
````

````python
# Medidas de dispersão
print("=== Medidas de Dispersão ===")
print(f"Desvio padrão: {df['nota'].std():.2f}")
print(f"Variância:     {df['nota'].var():.2f}")
print(f"Mínimo:        {df['nota'].min()}")
print(f"Máximo:        {df['nota'].max()}")
print(f"Amplitude:     {df['nota'].max() - df['nota'].min():.2f}")
````

````python
# NumPy — operações vetorizadas sobre arrays
notas = np.array(df["nota"])

print("Com NumPy:")
print(f"Média:         {np.mean(notas):.2f}")
print(f"Mediana:       {np.median(notas):.2f}")
print(f"Desvio padrão: {np.std(notas):.2f}")
print(f"Variância:     {np.var(notas):.2f}")

# Operações vetorizadas — aplicadas a todos os elementos de uma vez
print("\nNotas normalizadas (0 a 10 → 0 a 1):")
notas_norm = (notas - notas.min()) / (notas.max() - notas.min())
print(notas_norm.round(2))
````

````python
# describe() resume tudo de uma vez
print(df.describe())
````

## 6. Exemplo Prático — Fluxo Completo

Vamos simular o ciclo completo: carregar, limpar, manipular e analisar.

````python
import pandas as pd
import numpy as np

# 1. Dados brutos (simulando leitura de CSV)
dados = {
    "nome":   ["Ana", "Bruno", "Carlos", "Ana",  "Eva",   "Felipe"],
    "idade":  ["23",  "35",    "29",     "23",   None,    "31"],
    "nota":   [8.5,   7.0,    None,      8.5,    6.0,     9.5]
}

df = pd.DataFrame(dados)
print("--- Dados brutos ---")
print(df)
````

````python
# 2. Limpeza
df["nota"] = df["nota"].fillna(df["nota"].mean())  # preenche nota
df = df.dropna()                                    # remove linha sem idade
df = df.drop_duplicates()                           # remove duplicata
df["idade"] = df["idade"].astype(int)               # converte tipo

print("--- Após limpeza ---")
print(df)
````

````python
# 3. Manipulação
df["status"] = df["nota"].apply(lambda n: "Aprovado" if n >= 7 else "Reprovado")
df = df.sort_values(by="nota", ascending=False).reset_index(drop=True)

print("--- Após manipulação ---")
print(df)
````

````python
# 4. Análise
print("=== Estatísticas da Turma ===")
print(f"Média das notas:   {df['nota'].mean():.2f}")
print(f"Mediana das notas: {df['nota'].median():.2f}")
print(f"Desvio padrão:     {df['nota'].std():.2f}")
print()
print("Contagem por status:")
print(df["status"].value_counts())
print()
print("Média por status:")
print(df.groupby("status")["nota"].mean().round(2))
````
