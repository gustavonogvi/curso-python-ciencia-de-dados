
````python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Ana"],
    "idade": [23, None, 29, 23],
    "nota": [8.5, 7.0, None, 8.5]
}

df = pd.DataFrame(dados)
print("DataFrame original:")
print(df)

# Verificando nulos
print("\nValores nulos por coluna:")
print(df.isnull().sum())

# Preenchendo nota nula com a média
df["nota"] = df["nota"].fillna(df["nota"].mean())

# Removendo duplicados
df = df.drop_duplicates()

print("\nDataFrame após limpeza:")
print(df)
````

## 4. Manipulação de Dados

Agora vamos trabalhar com os dados.

### Seleção de colunas

```python
df["nome"]           # uma coluna
df[["nome", "nota"]] # múltiplas colunas
```

### Filtro de linhas

```python
df[df["nota"] >= 7]
```

### Ordenação

```python
df.sort_values(by="nota", ascending=False)
```

### Criando novas colunas

```python
df["aprovado"] = df["nota"] >= 7
```

### Agrupamento

```python
df.groupby("aprovado")["nota"].mean()
```

````python
import pandas as pd

dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Daniela", "Eduardo"],
    "idade": [23, 35, 29, 41, 27],
    "nota": [8.5, 6.0, 7.5, 9.0, 5.5]
}

df = pd.DataFrame(dados)

# Seleção de colunas
print("Coluna nome:")
print(df["nome"])

# Filtro
print("\nAlunos com nota >= 7:")
print(df[df["nota"] >= 7])

# Nova coluna
df["status"] = df["nota"].apply(lambda n: "Aprovado" if n >= 7 else "Reprovado")

# Ordenação
df_ordenado = df.sort_values(by="nota", ascending=False)
print("\nOrdenado por nota (maior -> menor):")
print(df_ordenado)

# Agrupamento
print("\nMédia por status:")
print(df.groupby("status")["nota"].mean())
````

## 5. Exemplo Prático

Vamos simular um fluxo completo de análise de dados.

````python
import pandas as pd

# 1. Criar os dados
dados = {
    "nome": ["Ana", "Bruno", "Carlos", "Ana"],
    "idade": [23, None, 29, 23],
    "nota": [8.5, 7.0, 9.2, 8.5]
}

df = pd.DataFrame(dados)
print("Original:")
print(df)

# 2. Limpeza
df = df.drop_duplicates()
df["idade"] = df["idade"].fillna(df["idade"].mean())

# 3. Manipulação
df["aprovado"] = df["nota"] >= 7
df = df.sort_values(by="nota", ascending=False)

# 4. Resultado final
print("\nResultado final:")
print(df)
````
